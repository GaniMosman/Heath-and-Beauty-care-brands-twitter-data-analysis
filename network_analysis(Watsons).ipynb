```python
import os
import sys
import json
from collections import Counter
import time
import math
import matplotlib.pyplot as plt 
import datetime
import csv
import networkx as nx
from operator import itemgetter
from wordcloud import WordCloud, STOPWORDS
import community

#read the node_list.csv
path1 = "output_watsons/node_list.csv"
with open(path1, 'r',encoding="utf8") as nodecsv:                 
    nodereader = csv.reader(nodecsv)  
    nodes = [n for n in nodereader][1:]


node_names = [n[1] for n in nodes]


#read the edge_list.csv
path2 = "output_watsons/edge_list.csv"
with open(path2, 'r') as edgecsv: 
    edgereader = csv.reader(edgecsv)      
    edges = [tuple(e) for e in edgereader][1:]

#create graph object
G = nx.Graph()

#add lists of nodes and edges 
G.add_edges_from(edges)
G.add_nodes_from(node_names)

#printing graph summary
print(nx.info(G))
```

```python
#plotting network
nx.draw(G)
plt.show()
```

```python
#finding density of network
density = nx.density(G)
print("Network density:", density)
```

```python
name = {}
id_ = {}
created_at={}
location = {}
description = {}
status_count = {}
friends_count = {}
followers_count = {}

for node in nodes:
    name[node[1]] = node[0]
    id_[node[1]] = node[2]
    created_at[node[1]] = node[3]
    location[node[1]] = node[4]
    description[node[1]] = node[5]
    status_count[node[1]] = node[6]
    friends_count[node[1]] = node[7]
    followers_count[node[1]] = node[8]
```

```python
nx.set_node_attributes(G, name = 'name', values = name)
nx.set_node_attributes(G, name = 'id', values = id_)
nx.set_node_attributes(G, name = 'created_at', values = created_at)
nx.set_node_attributes(G, name = 'location', values = location)
nx.set_node_attributes(G, name = 'description', values = description)
nx.set_node_attributes(G, name = 'status_count', values = status_count)
nx.set_node_attributes(G, name = 'friends_count', values = friends_count)
nx.set_node_attributes(G, name = 'followers_count', values = followers_count)
```

```python
for n in G.nodes():
    print(n, G.nodes[n]['name'])
```

```python
#finding degree of each nodes
person_dict = dict(G.degree(G.nodes()))
sorted_degree = sorted(person_dict.items(), key=itemgetter(1), reverse=True)
nx.set_node_attributes(G, name ='person_dict', values = person_dict)

print("Top 10 nodes by degree:")
for degree in sorted_degree[:10]:
    print(degree)
```

```python
#Creating community
communities = community.best_partition(G)

values = [communities.get(node) for node in G.nodes()]
```

```python
mod = community.modularity(communities, G)
```

```python
print(mod)
```

```python
nx.set_node_attributes(G, values = communities, name = "modularity")
```

```python
#Printing all members by each class
modularity = {}
for key,value in communities.items():
    if value not in modularity:
        modularity[value] = [key] 
    else:
        modularity[value].append(key) 

for key,value in modularity.items(): 
        print('Class '+str(key)+':', value)
        print("-----------------------------------------------------------------------------")
```

```python
#Determine which are the most central nodes in the each community(eigenvector)
nd_name_c1 = []
dg_c1 =[]

print("The Most Central nodes of Each Community based on Eigenvector Value")
large = 0
name = " "
print("Class                  Name                                  Eigenvector")
for k,v in modularity.items():  
    for x in v:
        if(G.nodes[x]['eigenvector'] >= large):
            large = G.nodes[x]['eigenvector']
            name = x
    nd_name_c1.append(name + "("+str(k)+")")
    dg_c1.append(large)
    print(k ,"                " , name + "                         ",large)
    large = 0
    name = " "
```
